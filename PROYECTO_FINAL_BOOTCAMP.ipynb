**Municipios con mayor y menor cobertura de acueducto**

#### **Top y Bottom 10 para cobertura de agua en zona urbana y rural**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top y bottom 10 en cobertura RURAL
top10_rural = df.sort_values('cobertura acueducto rural (%)', ascending=False).head(10)
bottom10_rural = df.sort_values('cobertura acueducto rural (%)', ascending=True).head(10)

# Top y bottom 10 en cobertura URBANA
top10_urbano = df.sort_values('cobertura acueducto urbano (%)', ascending=False).head(10)
bottom10_urbano = df.sort_values('cobertura acueducto urbano (%)', ascending=True).head(10)

# Gr√°ficos
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.barplot(data=top10_rural, y='municipio', x='cobertura acueducto rural (%)', ax=axes[0, 0],
            hue='municipio', palette='Greens', legend=False)
axes[0, 0].set_title('Top 10 - Cobertura Acueducto Rural')

sns.barplot(data=bottom10_rural, y='municipio', x='cobertura acueducto rural (%)', ax=axes[0, 1],
            hue='municipio', palette='Reds', legend=False)
axes[0, 1].set_title('Bottom 10 - Cobertura Acueducto Rural')

sns.barplot(data=top10_urbano, y='municipio', x='cobertura acueducto urbano (%)', ax=axes[1, 0],
            hue='municipio', palette='Blues', legend=False)
axes[1, 0].set_title('Top 10 - Cobertura Acueducto Urbano')

sns.barplot(data=bottom10_urbano, y='municipio', x='cobertura acueducto urbano (%)', ax=axes[1, 1],
            hue='municipio', palette='Oranges', legend=False)
axes[1, 1].set_title('Bottom 10 - Cobertura Acueducto Urbano')

# Ajustes finales
for ax in axes.flat:
    ax.set_ylabel('Cobertura (%)')
    ax.set_xlabel('')
    ax.tick_params(axis='y', labelsize=8)

plt.tight_layout()
plt.show()
```

#### **Clasificaci√≥n del IRCA MIN.SALUD sobre la calidad de agua**

```python
# Instalar librer√≠as si es necesario
!pip install pandas matplotlib seaborn openpyxl

# Importar
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files

# Configuraci√≥n
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10,6)

# Clasificaci√≥n IRCA
def clasificar_irca(irca):
    if irca <= 5:
        return 'Sin riesgo'
    elif irca <= 14:
        return 'Riesgo bajo'
    elif irca <= 35:
        return 'Riesgo medio'
    elif irca <= 80:
        return 'Riesgo alto'
    else:
        return 'Inviable sanitariamente'

df['nivel_irca'] = df['calidad del agua urbana - irca (%)'].apply(clasificar_irca)

# Paleta sem√°foro
colores = {
    'Sin riesgo': '#2ECC71',
    'Riesgo bajo': '#F1C40F',
    'Riesgo medio': '#E67E22',
    'Riesgo alto': '#E74C3C',
    'Inviable sanitariamente': '#8E44AD'
}

# Orden de categor√≠as y conteo seguro
orden = ['Sin riesgo', 'Riesgo bajo', 'Riesgo medio', 'Riesgo alto', 'Inviable sanitariamente']
conteo_series = df['nivel_irca'].value_counts().reindex(orden).fillna(0)
conteo_df = conteo_series.reset_index()
conteo_df.columns = ['nivel_irca', 'cantidad']
conteo_df['cantidad'] = conteo_df['cantidad'].astype(int)  # evitar NaN -> int error

# Gr√°fico
plt.figure()
sns.barplot(data=conteo_df, x='nivel_irca', y='cantidad', hue='nivel_irca',
            palette=colores, dodge=False, legend=False)

plt.title('Distribuci√≥n de Municipios seg√∫n Nivel de Riesgo IRCA')
plt.xlabel('Nivel de Riesgo')
plt.ylabel('N√∫mero de Municipios')

# Etiquetas encima
for i, row in conteo_df.iterrows():
    plt.text(i, row['cantidad'] + 0.5, str(row['cantidad']), ha='center', va='bottom', fontsize=12)

plt.tight_layout()
plt.show()
```

| >35 - 80       | Riesgo alto            |

#### **Gr√°fico sobre la brecha del acueducto en 15 municipios con mayor diferencia absoluta entre rural y urbano**

```python
# Tomamos solo 15 municipios con mayor diferencia absoluta entre rural y urbano
df['diferencia_absoluta'] = abs(df['cobertura acueducto urbano (%)'] - df['cobertura acueducto rural (%)'])
top_diff = df.sort_values(by='diferencia_absoluta', ascending=False).head(15)

# Formato largo (long format) para graficar
df_long = top_diff.melt(id_vars='municipio',
                        value_vars=['cobertura acueducto urbano (%)', 'cobertura acueducto rural (%)'],
                        var_name='tipo',
                        value_name='cobertura')

# Gr√°fico de barras comparativas
plt.figure()
sns.barplot(data=df_long, x='cobertura', y='municipio', hue='tipo', palette='Paired')
plt.title('Comparaci√≥n de Cobertura de Acueducto: Urbano vs Rural (Top 15 Desigualdad)')
plt.xlabel('Cobertura (%)')
plt.ylabel('Municipio')
plt.legend(title='Zona')
plt.tight_layout()
plt.show()
```

**Variables con una correlaci√≥n Fuertes (|r| > 0.5)**

```python
# 1. Seleccionar variables clave relacionadas con acceso y calidad
variables_clave = [
    'cobertura acueducto urbano (%)',
    'cobertura acueducto rural (%)',
    'cobertura alcantarillado urbano (%)',
    'cobertura aseo urbano (%)',
    'calidad del agua urbana - irca (%)',
    'porcentaje de aguas residuales tratadas (%)',
    'poblaci√≥n urbana (n√∫mero de habitantes)',
    'poblaci√≥n rural (n√∫mero de habitantes)'
]

# 2. Convertir variables cualitativas a num√©ricas
df['disposici√≥n_final'] = df['disposici√≥n final adecuada (s√≠/no)'].map({'s√≠': 1, 'no': 0})
df['aprovechamiento_viable'] = df['aprovechamiento viable (s√≠/no)'].map({'s√≠': 1, 'no': 0})

# 3. A√±adir columnas binarias al an√°lisis
variables_clave += ['disposici√≥n_final', 'aprovechamiento_viable']

# 4. Calcular matriz de correlaci√≥n focal
correlacion_focal = df[variables_clave].corr()

# 5. Filtrar correlaciones fuertes (mayores a 0.5 o menores a -0.5)
import numpy as np
mascaras_fuertes = (abs(correlacion_focal) > 0.5) & (abs(correlacion_focal) < 1)
correlaciones_fuertes = correlacion_focal.where(mascaras_fuertes).dropna(how='all').dropna(axis=1, how='all')

# 6. Mostrar correlaciones focales fuertes
print("üîç Correlaciones focales fuertes (|r| > 0.5):")
print(correlaciones_fuertes)

# 7. Heatmap compacto y legible
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7, 6))  # m√°s compacto
sns.heatmap(
    correlaciones_fuertes,
    annot=True,
    fmt=".2f",
    cmap='RdBu',
    center=0,
    annot_kws={"size": 8},        # texto dentro del mapa m√°s peque√±o
    cbar_kws={"shrink": 0.8}
)

plt.xticks(rotation=45, ha='right', fontsize=8)  # ejes m√°s peque√±os
plt.yticks(rotation=0, fontsize=8)
plt.title('Mapa de Correlaciones: Acceso y Calidad del Agua', fontsize=10)
plt.tight_layout()
plt.show()
```


##### **Interpretaci√≥n Detallada: Correlaciones Focales Fuertes (|r| > 0.5)**

##### **1.Cobertura acueducto urbano (%) ‚Üî Cobertura alcantarillado urbano (%)**

##### **r = 0.71** (correlaci√≥n positiva fuerte)  
Existe una relaci√≥n fuerte entre la cobertura de acueducto urbano y la de  alcantarillado urbano. Esto significa que los municipios que tienen una mayor proporci√≥n de viviendas urbanas con acceso a acueducto tambi√©n tienden a tener alta cobertura en alcantarillado.

##### **2.Cobertura alcantarillado urbano (%) ‚Üî Cobertura aseo urbano (%)**

##### **r = 0.53** (correlaci√≥n positiva moderada)  
Existe una relaci√≥n moderada entre la cobertura de alcantarillado y la del servicio de aseo en zona urbana. Es decir, donde hay mejores condiciones sanitarias de infraestructura, tambi√©n hay mejor cobertura del servicio de recolecci√≥n de residuos s√≥lidos

##### **3.Poblaci√≥n urbana ‚Üî Poblaci√≥n rural**

##### **r = 0.94** (correlaci√≥n positiva muy fuerte)  
Este resultado muestra que los municipios con mucha poblaci√≥n urbana tambi√©n suelen tener mucha poblaci√≥n rural. En otras palabras, los municipios con mayor n√∫mero de habitantes en general son grandes tanto en lo urbano como en lo rural.

```python
### Cantidad de proyectos priorizados en Antioquia para mejorar la accesibilidad al servicio de agua

proyectos_por_municipio = df_antioquia['MUNICIPIO(S)'].value_counts().reset_index()
proyectos_por_municipio.columns = ['MUNICIPIO', 'CANTIDAD DE PROYECTOS']
proyectos_por_municipio.head(15)
```

```python
# Contar cantidad de proyectos por tipo de DESTINO
conteo_destino = df_antioquia['DESTINO'].value_counts()

# Calcular porcentaje
porcentaje_destino = (conteo_destino / conteo_destino.sum()) * 100

# Crear un DataFrame bonito
df_porcentaje_destino = pd.DataFrame({
    'DESTINO': porcentaje_destino.index,
    'CANTIDAD': conteo_destino.values,
    'PORCENTAJE': porcentaje_destino.values.round(1)
})

df_porcentaje_destino.head()
```
